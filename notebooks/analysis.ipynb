# Analysis 2

```python
idx = 0
lab_id = df.iloc[idx]['lab_id']
video_id = df.iloc[idx]['video_id']

file_path = os.path.join(path, 'train_tracking', lab_id, str(video_id)+'.parquet')
temp = pd.read_parquet(file_path)

print(temp.shape)
print(temp.bodypart.nunique())
print(temp.bodypart.unique())
temp.head()
```

```python
print(temp.video_frame.nunique())
print(temp.video_frame.nunique()/30) # 30 frame per second
```

```python
frane_no = 20
from collections import Counter
print(len(temp.loc[temp['video_frame']==frane_no]['bodypart'].to_list()))
print(Counter(temp.loc[temp['video_frame']==frane_no]['bodypart'].to_list()))
temp.loc[temp['video_frame']==frane_no]
```
